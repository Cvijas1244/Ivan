```python
# Cs 2
import pandas as pd
import yfinance as yf
from prophet import Prophet
```

```python
# Cs 3
ticker = "SOUN"

start_date = "2020-01-01"
end_date = "2024-12-24"

data = yf.download(ticker, start=start_date, end=end_date)

data
```

```python
data = data.reset_index()
```

```python
# Cs 4
sub_data = data[["Date", "Adj Close"]]

sub_data
```

```python
# Cs 5
# Renaming the columns
sub_data.columns = ["ds", "y"]

sub_data
```

```python
# Cs 6
# Starting an instance of prophet
model = Prophet()
```

```python
# Code snippet 7
# Training the model using Prophet time series algo
model.fit(sub_data[['ds', 'y']])
```

```python
# Code snippet 8
# Preparing a 30 day future dataframe
future = model.make_future_dataframe(periods=30)
```

```python
# Code snippet 9
# Predicting the stock prices for next 30 periods
forecast = model.predict(future)
```

```python
# Code snippet 10
# Preparing a smaller forecast columns output
forecast_small = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]

forecast_small.tail(30)
```
